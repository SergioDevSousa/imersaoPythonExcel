```python
# importa pandas
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
```

```python
# print da tabela inteira dataframe principal
df_principal = pd.read_excel("/content/Imersão Python - Aula 02.xlsx", sheet_name="Principal")
df_principal
```

```python
# print dataframe total de açoes
df_Total_de_acoes = pd.read_excel("/content/Imersão Python - Aula 02.xlsx", sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
# print dataframe ticker
df_Ticker = pd.read_excel("/content/Imersão Python - Aula 02.xlsx", sheet_name="Ticker")
df_Ticker
```

```python
# print dataframe chatgpt
df_Chatgpt = pd.read_excel("/content/Imersão Python - Aula 02.xlsx", sheet_name="Chatgpt")
df_Chatgpt
```

```python
# leitura das colunas que iremos trabalhar
df_principal = df_principal [['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
# correção e alteração dos nomes das colunas com caracteres
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
# calculo da coluna valor percentual: =D2/100 e o calculo da coluna valor inicial: =C2/(L2+1) do google sheets ao python
df_principal['var_pct']=df_principal['var_dia_pct'] /100
df_principal['valor_inicial']=df_principal['valor_final']/(df_principal['var_pct']+1)
df_principal
```

```python
# comando merge (juntar) ele associa o dataframe total de ações para o dataframe principal como base da coluna ativo
df_principal = df_principal.merge (df_Total_de_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
# altera o nome da coluna em uma tabela
df_principal = df_principal.rename(columns={'Código':'codigo','Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

```python
# comando drop - exclui a coluna de uma tabela
df_principal =  df_principal.drop(columns=['codigo'])
df_principal
```

```python
# calculo no googlesheets criando a coluna var_real e seus valores =(C2-M2)*N2)
df_principal['var_real']=(df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['qtd_teorica']
df_principal
```

```python
# formata pra 2 casas decimais tipo flow
pd.options.display.float_format = '{:.2f}'.format
```

```python
# imprime dataframe
df_principal
```

```python
# transformar qtd_teorica (float) para (inteiro)
df_principal['qtd_teorica'] = df_principal['qtd_teorica'].astype(int)
df_principal
```

```python
# codigo extraido do chatGPT
df_principal['resultado'] = df_principal['var_real'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
# importar dados do dataframe ticker para o dataframe principal excluindo a coluna ticker
df_principal = df_principal.merge (df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal =  df_principal.drop(columns=['Ticker'])
df_principal
```

```python
# importa o restante das coluna, alem de exclui a coluna nome da empresa
df_principal = df_principal.merge (df_Chatgpt , left_on='Nome', right_on='Nome da empresa', how='left')
df_principal =  df_principal.drop(columns=['Nome da empresa'])
df_principal
```

```python
# altera Idade (anos) por idade_anos
df_principal = df_principal.rename(columns={'Idade (anos)':'idade_anos'}).copy()
df_principal
```

```python
df_principal['faixa'] = df_principal['idade_anos'].apply(lambda x: '+100' if x > 100 else ('ATÉ 50' if x < 50 else '50 ATÉ 100'))
df_principal
```

```python
# Calcula os valores desejados
maior = df_principal['var_real'].max()
menor = df_principal['var_real'].min()
media = df_principal['var_real'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['var_real'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['var_real'].mean()

# Imprime os resultados
print("Maior\t", f"R$ {maior:,.2f}")
print("Menor\t", f"R$ {menor:,.2f}")
print("Média\t", f"R$ {media:,.2f}")
print("Média subiu\t", f"R$ {media_subiu:,.2f}")
print("Média desceu\t", f"R$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_seguimento = df_principal_subiu.groupby('Segmento')['var_real'].sum().reset_index()
df_analise_seguimento
```

```python
fig = px.pie(df_analise_seguimento, 'Segmento', 'var_real',  title='Variação por segmento')

# Personalizar a legenda dentro de cada fatia da pizza
fig.update_traces(textinfo='label+percent', hovertemplate='Segmento: %{label}<br>Valor: R$ %{value:,.2f}', textposition='inside')

fig.show()
```
